**14번**

**답** 이진검색 트리와 마찬가지로 어떤 키의 직후 원소는 해당 키의 오른쪽 서브트리에서 가장 작은 값이다. B-트리 구조상 오른쪽 서브트리에서 왼쪽 자식을 타고 끝까지 내려가면 반드시 리프노드의 첫 번째 원소에 도달하게 됩니다.

---

**16번**

**답** B-트리에서 키의 개수가 N개라면, 외부노드의 개수는 항상 N+1개입니다, 키 자체의 분포는 삽입 순서에 따라 다르지만, 내부 노드의 키들은 리프 노드들 사이를 구분짓는 가이드 역할을 수행합니다.

---

**17번**

**답** 없다. 두 최소량의 키와 부모의 키 1개를 합쳐도 최대 키 수(M-1)를 넘지 않도록 설계되어 있다.
